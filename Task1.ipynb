## HANDLING MISSING VALUES
## IMPUTATION METHOD
## syntax:dataframe.fillna(value="any value",inplace=True)

```python
# filling the missing values in the director column
netflix_shows['director'].fillna(value="Rajiv Chilaka",inplace=True)
```

```python
# filling the missing values in cast column 
netflix_shows['cast'].fillna(value="David Attenborough",inplace=True)
```

```python
# filling the missing values in country column
netflix_shows['country'].fillna(value="United States",inplace=True)
```

```python
# filling the missing values in date_added column
netflix_shows['date_added'].fillna(value="January 1, 2020",inplace=True)
```

```python
# filling the missing values in the rating column
netflix_shows['rating'].fillna(value="TV-MA",inplace=True)
```

## Note:In the above i have replaced with the specific values because these are the most or top occuring values so i used them to replace the null values


```python
# rechecking the null values
netflix_shows.isnull().sum()
```
